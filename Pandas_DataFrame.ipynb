```python
#Step 1: Import the Pandas Library
#Explanation: This step imports the pandas library, which is used for creating and analyzing DataFrames.

import pandas as pd
```

```python
#Step 2: Create a DataFrame Using the Student Data
#Explanation: This step creates a pandas DataFrame using the given student dataset.

data = {
    "StudentID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ivan","Julia"],
    "Age": [20,21,19,22,20,21,23,19,22,20],
    "Marks": [88,75,92,65,78,85,70,95,60,80],
    "Grade": ["A","B","A","C","B","A","B","A","C","B"]
}

df = pd.DataFrame(data)

df
```

```python
#Step 3: Save the DataFrame as a CSV File
#Explanation: This step saves the DataFrame into a CSV file named students.csv.

df.to_csv("students.csv", index=False)

print("CSV file saved successfully.")
```

```python
#Step 4: Reload the CSV File into a New DataFrame
#Explanation: This step reads the saved CSV file and loads it into a new DataFrame.

df_new = pd.read_csv("students.csv")
df_new
```

```python
#Step 5: Display the First 3 Rows and Display Statistical Summary etc..,.
#Explanation: This step explores the dataset by displaying the first three rows, statistical summary, shape, data types, and overall DataFrame information using head(), describe(), shape, dtypes, and info().

df_new.head(3)
df_new.describe()
df_new.shape
df_new.dtypes
df_new.info()
```

```python
#Complete Program code in one block

# Import pandas library
import pandas as pd

# Create student data
data = {
    "StudentID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ivan","Julia"],
    "Age": [20,21,19,22,20,21,23,19,22,20],
    "Marks": [88,75,92,65,78,85,70,95,60,80],
    "Grade": ["A","B","A","C","B","A","B","A","C","B"]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame as CSV
df.to_csv("students.csv", index=False)

# Reload CSV file
df_new = pd.read_csv("students.csv")

# Display first 3 rows
print("First 3 Rows:")
print(df_new.head(3))

# Display statistical summary
print("\nStatistical Summary:")
print(df_new.describe())

# Display shape of DataFrame
print("\nShape of DataFrame:")
print(df_new.shape)

# Display data types
print("\nData Types:")
print(df_new.dtypes)

# Display complete information
print("\nDataFrame Info:")
df_new.info()
```
