```python
# Data Merging and Lookup Operations with Pandas

#Student Name: Nanthini M

import pandas as pd
import numpy as np
```

# TASK 1: Data Preparation and Missing Value Handling

```python

# Students DataFrame
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

students = pd.DataFrame(students_data)


# Enrollments DataFrame
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

enrollments = pd.DataFrame(enrollments_data)


# Scores DataFrame
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

scores = pd.DataFrame(scores_data)


print("Students DataFrame:")
display(students)
```

```python
# Null Value Analysis
print("Null Value Analysis")

null_counts = students.isnull().sum()
null_percent = (students.isnull().sum() / len(students)) * 100

for col in students.columns:
    print(f"Column: {col}, Nulls: {null_counts[col]} ({null_percent[col]:.2f}%)")
```

```python
# Handle Missing Values
# Fill missing city
students['city'] = students['city'].fillna('Unknown')

# Drop rows where name is missing
students_clean = students.dropna(subset=['name'])

print("Cleaned Students DataFrame")
display(students_clean)
```

## TASK 2: Multiple Join Operations

```python
inner_join = pd.merge(students_clean, enrollments, on='student_id', how='inner')

print("Inner Join Result:")
display(inner_join)

print("Number of students:", len(inner_join))
```

```python
left_join = pd.merge(students_clean, enrollments, on='student_id', how='left')

print("Left Join Result")
display(left_join)

print("Total rows:", len(left_join))
```

```python
right_join = pd.merge(students_clean, enrollments, on='student_id', how='right')

display(right_join)

print("Total rows:", len(right_join))
```

```python
outer_join = pd.merge(students_clean, enrollments, on='student_id', how='outer', indicator=True)

display(outer_join)

print("Total rows:", len(outer_join))
```

```python
missing_rows = outer_join[(outer_join['name'].isnull()) | (outer_join['course_name'].isnull())]

display(missing_rows)
```

## TASK 3: Lookup Operations and Automation

```python
score_dict = dict(zip(scores.student_id, scores.exam_score))

students_clean['exam_score'] = students_clean['student_id'].map(score_dict)

display(students_clean[['student_id','name','exam_score']])
```

```python
merge_scores = pd.merge(students_clean.drop(columns=['exam_score'], errors='ignore'),
                        scores,
                        on='student_id',
                        how='left')

display(merge_scores[['student_id','name','exam_score']])
```

```python
def auto_merge(df1, df2, join_type, key_column):

    merged = pd.merge(df1, df2, how=join_type, on=key_column)

    return {
        "result_df": merged,
        "row_count": len(merged),
        "join_type": join_type
    }
```

```python
test1 = auto_merge(students_clean, enrollments, "inner", "student_id")
test2 = auto_merge(students_clean, enrollments, "left", "student_id")

print("Join Type:", test1['join_type'])
print("Rows:", test1['row_count'])
display(test1['result_df'].head())

print("Join Type:", test2['join_type'])
print("Rows:", test2['row_count'])
display(test2['result_df'].head())
```
