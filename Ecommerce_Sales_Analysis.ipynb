```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

print(df.head(10))
print("\nDataset shape:", df.shape)
```

```python
# Total Sales Amount for Each Region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()

print(region_sales)
```

```python
# Number of Transactions per Product Category
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

print(category_transactions)
```

```python
# Average Sales Amount per Salesperson
avg_sales_salesperson = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print(avg_sales_salesperson)
```

```python
# Identify the Top Performing Region
region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)

print(region_sales_sorted)
```

```python
# Task 2: Multi-Column Grouping and Multiple Aggregations
# Total Sales by Region and Category
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

print(region_category_sales)
```

```python
# Salesperson Performance Metrics
salesperson_performance = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

print(salesperson_performance)

# Identify Top Salesperson
top_salesperson = salesperson_performance.sort_values(by='sum', ascending=False)

print(top_salesperson)

# Category with Maximum Revenue
category_sales = df.groupby('category')['sales_amount'].sum()

max_category = category_sales.idxmax()

print("Category with highest revenue:", max_category)

```

```python
# Task 3: Custom Aggregation and Complete Sales Report
# Custom Aggregation Function
def sales_range(x):
    return x.max() - x.min()

# Apply Custom Aggregation to Regions
region_analysis = df.groupby('region')['sales_amount'].agg(['sum','mean','min','max',sales_range]).reset_index()

print(region_analysis)

# Final Regional Sales Summary Report
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print(final_report)

# Multi-Level Columns
df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
})
```
