```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# Preview data
tips.head()
```

```python
# Task 1 — Matplotlib Bar Chart
tips_sum = tips.groupby("day")["tip"].sum()
```

```python
# Plot Bar Chart
plt.figure(figsize=(8, 5))

tips_sum.plot(kind='bar')

plt.title("Total Tips Collected per Day")
plt.xlabel("Day")
plt.ylabel("Total Tip Amount")

plt.show()
```

```python
# Task 2 — Seaborn Histogram
plt.figure(figsize=(8, 5))

sns.histplot(tips["tip"], bins=20, kde=True)

plt.title("Distribution of Tip Amounts")

plt.show()
```

Most of the tips are concentrated between $2 and $4, indicating that customers generally leave moderate tip amounts. The distribution appears slightly right-skewed, meaning there are fewer high-value tips. The KDE curve helps visualize the smooth distribution and overall trend of tipping behavior.

```python
# Task 3 — Seaborn Scatter Plot
plt.figure(figsize=(8, 5))

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

plt.title("Total Bill vs Tip Amount")

plt.show()
```

There is a positive relationship between total bill and tip amount, meaning higher bills generally result in higher tips. However, the spread of points shows some variation, indicating that tipping behavior is not strictly proportional.
